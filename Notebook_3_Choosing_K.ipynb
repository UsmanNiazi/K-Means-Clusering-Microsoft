# How to Choose the Number of Clusters

## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

```python
data = pd.read_csv('Notebook_3.csv')

data
```

## Plot the data

```python
plt.scatter(data['Longitude'], data['Latitude'])
plt.xlim(-180,180)
plt.ylim(-90, 90)
plt.show()
```

## Select the features

```python
x = data.iloc[:,1:]
x
```

## Clustering

```python
km = KMeans(4)

km.fit(x)
```

### Clustering results

```python
df = km.fit_predict(x)
```

```python
newdf = data.copy()

newdf['pred'] = df

newdf
```

Plot the data once again but separate the data by the clusters we defined.

## Plot the data

```python
plt.scatter(newdf['Longitude'],newdf['Latitude'], c = newdf['pred'], cmap = 'rainbow')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```

## Selecting the number of clusters

### WCSS

```python
wcss = []

test = range(1,10)

for i in test:
    kmm = KMeans(i)
    kmm.fit(x)
    wcss_j = kmm.inertia_
    wcss.append(wcss_j)

wcss
```

```python
km.inertia_
```

### The Elbow Method

```python
nc = range(1,10)
plt.plot(nc,wcss)
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Clusters Sum of squares')

```
